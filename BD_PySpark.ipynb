### TASK-1:
Generate 100 random numbers in range 0 to 10 using numpy randint function with the seed set to 10. Create a RDD using the parallelize function using data generated in previous step. Calculate the frequency of each number (0 - 10) using appropriate function of RDD.

```python
pip install pyspark
```

```python
from pyspark import SparkContext
import numpy as np

# Initialize a SparkContext
sc = SparkContext("local", "RandomNumbersRDD")

# Set seed for reproducibility
np.random.seed(10)

random_numbers = np.random.randint(0, 11, 100)

# Create an RDD using parallelize function
random_numbers_rdd = sc.parallelize(random_numbers)

# Calculate the frequency of each number
frequency_count = random_numbers_rdd.countByValue()

print("Generated Random Numbers:")
print(random_numbers)

print("\nFrequency of Each Number:")
for number, count in sorted(frequency_count.items()):
    print(f"Number {number}: {count} occurrences")

# Stop the SparkContext
sc.stop()
```

### TASK-2:
In this task you will calculate the frequency of each word in text8 dataset. Create a RDD using the text8 dataset. Use appropriate functions of the RDD to get the word frequencies. Filter the RDD using appropriate function to get the frequencies of words containing the letter ’a’.

```python
# Initialize a SparkContext
sc = SparkContext("local", "WordFrequency")

text_rdd = sc.textFile("/content/text8")

# Split each line into words and flatten the list of words
words_rdd = text_rdd.flatMap(lambda line: line.split())

# Map each word to a key-value pair with the word as the key and 1 as the value
word_count_rdd = words_rdd.map(lambda word: (word, 1))

# Reduce by key to get the frequency of each word
word_frequencies = word_count_rdd.reduceByKey(lambda x, y: x + y)

print("Word Frequencies:")
for word, count in word_frequencies.collect():
    print(f"Word '{word}': {count} occurrences")

# Filter the RDD to get frequencies of words containing the letter 'a'
words_with_a = word_frequencies.filter(lambda x: 'a' in x[0])

# Print the frequencies of words containing the letter 'a'
print("\nWord Frequencies with 'a':")
for word, count in words_with_a.collect():
    print(f"Word '{word}': {count} occurrences")

# Stop the SparkContext
sc.stop()
```

##DataFrame Task:

Create a Spark dataframe using the iris json data. Calculate Pearson Correlation between the columns petalLength and petalWidth using the appropriate dataframe API. Show the columns sepalLength, sepalWidth and species for the rows of data that has petalLength greater than or equal to 1.4 using the appropriate dataframe API.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, corr

# Initialize a Spark session
spark = SparkSession.builder.appName("IrisDataFrame").getOrCreate()

iris_df = spark.read.json("/content/iris.json")

# Show the DataFrame schema
iris_df.printSchema()

# Calculate Pearson Correlation between petalLength and petalWidth
correlation = iris_df.select(corr("petalLength", "petalWidth").alias("correlation")).collect()[0]["correlation"]
print(f"\nPearson Correlation between petalLength and petalWidth: {correlation}")

# Show columns sepalLength, sepalWidth, and species for rows where petalLength is greater than or equal to 1.4
filtered_data = iris_df.filter(col("petalLength") >= 1.4).select("sepalLength", "sepalWidth", "species")
print("\nFiltered Data:")
filtered_data.show()

spark.stop()
```

```python
# Create or get SparkContext
sc = SparkContext.getOrCreate()
# Load text file from Google Drive
rdd_google = spark.sparkContext.textFile("/content/drive/MyDrive/Dataset/sample.txt")

# Define path for text file stored in Databricks File System (DBFS)
path_dbfs = 'dbfs:/FileStore/shared_uploads/202318054@daiict.ac.in/sample-2.txt'
# Load text file from DBFS
rdd_dbfs = spark.sparkContext.textFile(path_dbfs)
# Collect the contents of the file into a list
file = rdd_dbfs.collect()
```

```python
# Import regular expression module
import re
# Define pattern for splitting words
pattern = r'[\n\s!@#$%^&*\(\)\{\}\[\]~`/\?<>,\.]+'
# Split the file contents into words using the defined pattern
words = re.split(pattern, str(file))
words
```

```python
# Create a list of tuples containing each word and its count as 1
word_count = list(map(lambda x: (x, 1), words))
word_count
```

```python
word_counts = {}
# Count the occurrences of each word
for word in words:
    word = word.lower()
    word_counts[word] = word_counts.get(word, 0) + 1

# Sort the word counts in descending order based on the count
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

print(f'Total number of words: {sum(word_counts.values())}\n')
print('List of Words and their Counts (sorted by count in descending order):\n')
for word, count in sorted_word_counts:
    print(f'{word}: {count}')
```
